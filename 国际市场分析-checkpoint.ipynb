```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

# 设置随机种子保证可重现性
np.random.seed(42)

# 基础参数设置
start_date = '2020-01-01'
end_date = '2023-12-31'

# 扩展国家列表（新增10个国家）
countries = [
    '美国', '中国', '日本', '德国', '巴西', '印度', '澳大利亚',
    '英国', '法国', '加拿大', '墨西哥', '韩国', '俄罗斯', '南非',
    '新加坡', '沙特阿拉伯', '土耳其', '越南', '印度尼西亚', '泰国'
]

# 详细产品分类（主类别和子类别）
categories = {
    '电子产品': ['智能手机', '笔记本电脑', '平板电脑', '智能手表', '耳机'],
    '服装': ['男装', '女装', '童装', '运动装', '鞋类'],
    '家居用品': ['家具', '厨房用品', '装饰品', '床上用品', '照明'],
    '食品': ['零食', '饮料', '冷冻食品', '调味品', '健康食品'],
    '美妆个护': ['护肤品', '彩妆', '香水', '护发产品', '身体护理']
}

# 货币代码
currency_codes = {
    '美国': 'USD', '中国': 'CNY', '日本': 'JPY', '德国': 'EUR', '巴西': 'BRL',
    '印度': 'INR', '澳大利亚': 'AUD', '英国': 'GBP', '法国': 'EUR', '加拿大': 'CAD',
    '墨西哥': 'MXN', '韩国': 'KRW', '俄罗斯': 'RUB', '南非': 'ZAR', '新加坡': 'SGD',
    '沙特阿拉伯': 'SAR', '土耳其': 'TRY', '越南': 'VND', '印度尼西亚': 'IDR', '泰国': 'THB'
}

# 国家分组（用于设置不同规则）
developed_countries = ['美国', '日本', '德国', '澳大利亚', '英国', '法国', '加拿大', '韩国', '新加坡']
emerging_countries = ['中国', '印度', '巴西', '墨西哥', '俄罗斯', '南非', '土耳其', '越南', '印度尼西亚', '泰国', '沙特阿拉伯']

# 生成日期范围（月度数据）
date_range = pd.date_range(start=start_date, end=end_date, freq='MS')

# 创建基础数据框架
df = pd.DataFrame()

for date in date_range:
    for country in countries:
        for main_category, subcategories in categories.items():
            for subcategory in subcategories:
                # 基础销售额（当地货币）
                base_sales = np.random.uniform(20000, 150000)
                
                # 时间趋势（每年增长）
                time_factor = (date.year - 2020) * 0.12
                
                # 季节性因素（Q4旺季，Q1淡季）
                if date.month in [10, 11, 12]:
                    seasonal_factor = np.random.uniform(1.3, 1.7)
                elif date.month in [1, 2, 3]:
                    seasonal_factor = np.random.uniform(0.6, 0.85)
                else:
                    seasonal_factor = np.random.uniform(0.9, 1.1)
                    
                # 国家增长因子（新兴市场增长更快）
                if country in emerging_countries:
                    country_factor = np.random.uniform(1.15, 1.4)
                else:
                    country_factor = np.random.uniform(0.95, 1.1)
                    
                # 类别差异因子
                if main_category == '电子产品':
                    category_factor = np.random.uniform(1.3, 1.6)
                elif main_category == '美妆个护':
                    category_factor = np.random.uniform(1.1, 1.4)
                elif main_category == '食品':
                    category_factor = np.random.uniform(0.9, 1.1)
                else:
                    category_factor = 1.0
                    
                # 子类别调整
                subcategory_factors = {
                    '智能手机': 1.8, '笔记本电脑': 1.5, '平板电脑': 1.3, 
                    '男装': 0.9, '女装': 1.4, '运动装': 1.2,
                    '家具': 0.8, '厨房用品': 1.1, '零食': 1.0, '饮料': 1.3,
                    '护肤品': 1.4, '彩妆': 1.6
                }
                sub_factor = subcategory_factors.get(subcategory, 1.0)
                
                # 计算最终销售额（当地货币）
                local_sales = base_sales * (1 + time_factor) * seasonal_factor * country_factor * category_factor * sub_factor
                
                # 添加随机噪声
                local_sales *= np.random.uniform(0.92, 1.08)
                
                # 汇率生成（对美元）
                if country == '美国':
                    exchange_rate = 1.0
                else:
                    # 基础汇率 + 随机波动 + 时间趋势
                    base_rates = {
                        '中国': 6.8, '日本': 110, '德国': 0.9, '巴西': 5.2, 
                        '印度': 75, '澳大利亚': 1.4, '英国': 0.78, '法国': 0.9,
                        '加拿大': 1.3, '墨西哥': 20, '韩国': 1200, '俄罗斯': 75,
                        '南非': 15, '新加坡': 1.35, '沙特阿拉伯': 3.75, '土耳其': 8.5,
                        '越南': 23000, '印度尼西亚': 14500, '泰国': 33
                    }
                    trend = (2023 - date.year) * np.random.uniform(-0.07, 0.07)
                    volatility = np.random.uniform(-0.15, 0.15)
                    exchange_rate = base_rates[country] * (1 + trend) * (1 + volatility)
                
                # 计算美元销售额
                usd_sales = local_sales / exchange_rate
                
                # 经济指标（模拟GDP增长率和通胀率）
                if country in ['中国', '印度', '越南']:
                    gdp_growth = np.random.uniform(0.06, 0.09)
                    inflation = np.random.uniform(0.03, 0.06)
                elif country in ['巴西', '土耳其', '阿根廷']:
                    gdp_growth = np.random.uniform(0.01, 0.03)
                    inflation = np.random.uniform(0.08, 0.15)
                else:
                    gdp_growth = np.random.uniform(0.02, 0.04)
                    inflation = np.random.uniform(0.01, 0.03)
                
                # ====== 新增成本因素 
                  # 产品成本（销售额的百分比）
                if main_category == '电子产品':
                    product_cost_rate = np.random.uniform(0.55, 0.65)
                elif main_category == '食品':
                    product_cost_rate = np.random.uniform(0.45, 0.55)
                else:
                    product_cost_rate = np.random.uniform(0.4, 0.5)
                
                # 运费（基于距离和产品类型）
                # 基准运费 + 国家距离因子 + 产品体积因子
                base_shipping = 500  # 基础运费
                
                # 国家距离因子（模拟）
                distance_factors = {
                    '美国': 1.0, '加拿大': 1.2, '墨西哥': 1.5,
                    '英国': 1.8, '德国': 1.9, '法国': 1.9,
                    '中国': 2.5, '日本': 2.7, '澳大利亚': 2.8,
                    '印度': 2.6, '巴西': 2.3, '俄罗斯': 2.4,
                    '南非': 2.7, '新加坡': 2.6, '沙特阿拉伯': 2.2,
                    '土耳其': 2.0, '越南': 2.7, '印度尼西亚': 2.8, '泰国': 2.7
                    
                }
                
                # 产品体积因子
                size_factors = {
                    '智能手机': 0.8, '笔记本电脑': 1.2, '平板电脑': 1.0,
                    '家具': 3.0, '厨房用品': 1.5, '饮料': 1.8
                }
                size_factor = size_factors.get(subcategory, 1.0)
                
                shipping_cost = base_shipping * distance_factors.get(country, 2.0) * size_factor
                
                # 关税（基于国家和产品类型）
                if country in ['美国', '加拿大', '墨西哥']:  # USMCA
                    tariff_rate = np.random.uniform(0.01, 0.05)
                elif country in ['德国', '法国', '英国']:  # 欧盟类似
                    tariff_rate = np.random.uniform(0.03, 0.07)
                elif country in emerging_countries:
                    tariff_rate = np.random.uniform(0.08, 0.15)
                else:
                    tariff_rate = np.random.uniform(0.05, 0.10)
                
                # 特殊产品关税
                if main_category == '电子产品':
                    tariff_rate *= np.random.uniform(1.1, 1.3)
                elif main_category == '食品':
                    tariff_rate *= np.random.uniform(0.8, 1.0)
                
                tariff_cost = usd_sales * tariff_rate
                
                # 计算总成本和利润
                product_cost = usd_sales * product_cost_rate
                total_cost = product_cost + shipping_cost + tariff_cost
                profit = usd_sales - total_cost
                profit_margin = profit / usd_sales if usd_sales > 0 else 0
                
                # 添加到数据框
                new_row = pd.DataFrame({
                    '日期': [date],
                    '国家': [country],
                    '主类别': [main_category],
                    '子类别': [subcategory],
                    '当地货币销售额': [local_sales],
                    '汇率': [exchange_rate],
                    '美元销售额': [usd_sales],
                    'GDP增长率': [gdp_growth],
                    '通胀率': [inflation],
                    '货币': [currency_codes[country]],
                    '产品成本率': [product_cost_rate],
                    '产品成本': [product_cost],
                    '运费': [shipping_cost],
                    '关税税率': [tariff_rate],
                    '关税成本': [tariff_cost],
                    '总成本': [total_cost],
                    '利润': [profit],
                    '利润率': [profit_margin]
                })
                df = pd.concat([df, new_row], ignore_index=True)

# 添加季度和年份列
df['季度'] = df['日期'].dt.quarter
df['年份'] = df['日期'].dt.year

# 保存到CSV
df.to_csv('extended_international_sales_data.csv', index=False)

print("扩展的模拟数据生成完成！包含 {} 行记录".format(len(df)))
print("国家数量: {}, 产品子类别: {}".format(len(countries), sum(len(v) for v in categories.values())))
```

```python
from scipy import stats

# 计算 Z-score
z_scores = stats.zscore(df[['美元销售额', '运费', '关税成本']])

# 设置阈值（例如，绝对值大于3的被认为是异常值）
threshold = 3
outliers = (abs(z_scores) > threshold).any(axis=1)

# 打印异常值记录
print(df[outliers])
df.loc[outliers, '美元销售额'] = df['美元销售额'].median()
df.loc[outliers, '运费'] = df['运费'].median()
df.loc[outliers, '关税成本'] = df['关税成本'].median()
```

```python

cost_analysis = df.groupby('国家').agg({
    '利润': 'sum',
    '运费': 'sum',
    '关税成本': 'sum',
    '美元销售额': 'sum'
}).reset_index()

# Step 3: 计算各项占比
cost_analysis['利润率'] = cost_analysis['利润'] / cost_analysis['美元销售额']
cost_analysis['运费占比'] = cost_analysis['运费'] / cost_analysis['美元销售额']
cost_analysis['关税占比'] = cost_analysis['关税成本'] / cost_analysis['美元销售额']

# 计算中位数
freight_median = np.median(cost_analysis['运费占比'] )
tariff_median = np.median(cost_analysis['关税占比'] )
lirun_median=np.median(cost_analysis['利润率'])

# 设置阈值
x_thres = freight_median * 1.2  # 略高于中位数
y_thres = tariff_median * 1.2   # 略高于中位数
# 假设利润率中位数作为阈值
profit_median = cost_analysis['利润率'].median()



# Step 5: 初始化区域列表
high_profit_zone = []
challenge_zone = []
opportunity_zone = []
stable_zone = []

# 修改分类输出，加入利润率标签
def get_profit_label(profit):
    return "🟢高利润" if profit >= profit_median else "🔴低利润"

# 在分类时加入利润标签
for _, row in cost_analysis.iterrows():
    country = row['国家']
    freight_ratio = row['运费占比']
    tariff_ratio = row['关税占比']
    profit_label = get_profit_label(row['利润率'])

    if freight_ratio < x_thres and tariff_ratio >= y_thres:
        high_profit_zone.append(f"{country} ({profit_label}) 运费:{freight_ratio:.1%}, 关税:{tariff_ratio:.1%}")
    elif freight_ratio >= x_thres and tariff_ratio >= y_thres:
        challenge_zone.append(f"{country} ({profit_label}) 运费:{freight_ratio:.1%}, 关税:{tariff_ratio:.1%}")
    elif freight_ratio >= x_thres and tariff_ratio < y_thres:
        opportunity_zone.append(f"{country} ({profit_label}) 运费:{freight_ratio:.1%}, 关税:{tariff_ratio:.1%}")
    else:
        stable_zone.append(f"{country} ({profit_label}) 运费:{freight_ratio:.1%}, 关税:{tariff_ratio:.1%}")

# Step 7: 输出结果
print("=====================================")
print("      各国家所属区域分类结果")
print("=====================================")

print("\n🟢 高利润区（低运费 + 高关税）:")
if high_profit_zone:
    for item in high_profit_zone:
        print(f"  • {item}")
else:
    print("  （无）")

print("\n🔴 挑战区（高运费 + 高关税）:")
if challenge_zone:
    for item in challenge_zone:
        print(f"  • {item}")
else:
    print("  （无）")

print("\n🟡 机会区（高运费 + 低关税）:")
if opportunity_zone:
    for item in opportunity_zone:
        print(f"  • {item}")
else:
    print("  （无）")

print("\n🔵 稳定区（低运费 + 低关税）:")
if stable_zone:
    for item in stable_zone:
        print(f"  • {item}")
else:
    print("  （无）")


```

```python
# ====================== 4. 产品定价弹性分析 ======================
# 模拟价格数据（原数据集中无价格，需扩展）
np.random.seed(42)
products = df['子类别'].unique()
price_elasticity = {}

for product in products:
    # 模拟基础价格
    base_price = np.random.uniform(50, 500)
    
    # 模拟价格变动影响
    price_changes = np.linspace(-0.2, 0.2, 5)  # -20%到+20%价格变动
    demand_changes = []
    
    for change in price_changes:
        # 不同产品价格弹性不同
        if '智能' in product:
            elasticity = -1.5  # 高弹性
        elif '服装' in product:
            elasticity = -2.0
        else:
            elasticity = -1.0
            
        demand_change = elasticity * change
        demand_changes.append(demand_change)
    
    price_elasticity[product] = {
        '价格变动': price_changes,
        '需求变动': demand_changes,
        '价格弹性': elasticity
    }

# 可视化价格弹性
elasticity_df = pd.DataFrame({
    '产品': list(price_elasticity.keys()),
    '弹性系数': [v['价格弹性'] for v in price_elasticity.values()]
}).sort_values('弹性系数')

plt.figure(figsize=(12, 8))
sns.barplot(x='弹性系数', y='产品', data=elasticity_df, palette='coolwarm')
plt.title('产品价格弹性系数', fontsize=16)
plt.xlabel('价格弹性系数 (负值表示需求随价格上涨而下降)')
plt.ylabel('')
plt.axvline(-1.0, color='red', ls='--', label='弹性需求分界线')
plt.legend()
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
价格弹性系数衡量了商品需求量对价格变动的敏感程度。通常，价格弹性系数大于1表示需求富有弹性，即价格的小幅变化会引起需求量的大幅变化；
价格弹性系数小于1表示需求缺乏弹性，即价格变化对需求量的影响较小


1.高弹性产品（蓝色区域）：
如智能手机、智能手表等电子产品和香水、彩妆等奢侈品。这些产品的价格弹性系数较高，表明消费者对价格变化较为敏感，价格的微小变动可能会导致需求量的显著变化。
2.中等弹性产品（浅蓝色区域）：
如护肤品、健康食品、调味品等。这类产品的价格弹性系数介于高弹性和低弹性之间，消费者对价格变化有一定的敏感度，但不会像高弹性产品那样剧烈反应。
3.低弹性产品（灰色区域）：
如床上用品、装饰品、厨房用品等家居用品。这些产品的价格弹性系数较低，表明消费者对价格变化不太敏感，即使价格有所变动，需求量的变化也相对较小。
4.极低弹性产品（橙色区域）：
如家具、鞋类、运动装、童装、女装等。这类产品的价格弹性系数最低，消费者对价格变化几乎不敏感，价格变动对需求量的影响非常有限。
```

```python
## ====================== 5. 产品盈利能力与成本结构 ======================
# 计算各产品成本占比
product_costs = df.groupby(['主类别', '子类别']).agg(
    sales=('美元销售额', 'sum'),
    cost=('总成本', 'sum'),
    profit=('利润', 'sum')
).reset_index()

product_costs['利润率'] = product_costs['profit'] / product_costs['sales'] * 100
product_costs['成本率'] = product_costs['cost'] / product_costs['sales'] * 100

# 成本结构分解
product_cost_breakdown = df.groupby(['主类别', '子类别']).agg(
    product_cost=('产品成本', 'sum'),
    shipping_cost=('运费', 'sum'),
    tariff_cost=('关税成本', 'sum')
).reset_index()

product_cost_breakdown['总成本'] = product_cost_breakdown[['product_cost', 'shipping_cost', 'tariff_cost']].sum(axis=1)
product_cost_breakdown['产品成本占比'] = product_cost_breakdown['product_cost'] / product_cost_breakdown['总成本'] * 100
product_cost_breakdown['运费占比'] = product_cost_breakdown['shipping_cost'] / product_cost_breakdown['总成本'] * 100
product_cost_breakdown['关税占比'] = product_cost_breakdown['tariff_cost'] / product_cost_breakdown['总成本'] * 100

# 合并数据
product_analysis = pd.merge(product_costs, product_cost_breakdown, on=['主类别', '子类别'])

# 可视化
plt.figure(figsize=(16, 10))

# 利润率 vs 成本率
plt.subplot(2, 2, 1)
sns.scatterplot(
    x='成本率',
    y='利润率',
    hue='主类别',
    size='sales',
    sizes=(50, 500),
    alpha=0.7,
    data=product_analysis
)
plt.title('产品利润率 vs 成本率')
plt.xlabel('成本率 (%)')
plt.ylabel('利润率 (%)')
plt.axhline(0, color='red', ls='--')
plt.grid(alpha=0.3)

# 成本结构堆叠图
plt.subplot(2, 2, 2)
cost_melt = product_analysis.melt(
    id_vars=['子类别'],
    value_vars=['产品成本占比', '运费占比', '关税占比'],
    var_name='成本类型',
    value_name='占比'
)

sns.barplot(
    x='占比',
    y='子类别',
    hue='成本类型',
    data=cost_melt,
    palette='Set2'
)
plt.title('产品类别成本结构')
plt.xlabel('成本占比 (%)')
plt.ylabel('')

# 成本效率分析 (离群值检测)
plt.subplot(2, 2, 3)
product_analysis['成本效率'] = product_analysis['sales'] / product_analysis['cost']
product_analysis['成本效率Z'] = zscore(product_analysis['成本效率'])

sns.barplot(
    x='成本效率',
    y='子类别',
    data=product_analysis.sort_values('成本效率', ascending=False),
    palette='viridis'
)
plt.title('产品成本效率 (销售额/总成本)')
plt.xlabel('成本效率')
plt.ylabel('')

# 成本结构气泡图
plt.subplot(2, 2, 4)
sns.scatterplot(
    x='运费占比',
    y='关税占比',
    size='利润率',
    hue='主类别',
    sizes=(50, 300),
    alpha=0.7,
    data=product_analysis
)
plt.title('产品运费与关税占比')
plt.xlabel('运费占比 (%)')
plt.ylabel('关税占比 (%)')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('产品成本结构.png', dpi=300)
plt.show()
```

```python
## ====================== 6. 产品创新与增长机会 ======================

# 1. 计算每年销售额
yearly_sales = df.groupby(['年份', '主类别', '子类别'])['美元销售额'].sum().reset_index()

# 2. 创建透视表：行为年份，列为 (主类别, 子类别)，值为销售额
sales_pivot = yearly_sales.pivot_table(
    index='年份',
    columns=['主类别', '子类别'],
    values='美元销售额',
    fill_value=0
)

# 3. 计算同比增长率（2023 相比 2022）
growth_pivot = sales_pivot.pct_change().loc[2023]  # 2023年增长率
product_growth_trend = growth_pivot.reset_index()
product_growth_trend.columns = ['主类别', '子类别', '增长率']

# 4. 筛选高增长产品（增长率 > 30%）
high_growth = product_growth_trend[product_growth_trend['增长率'] > 0.3].copy()
high_growth = high_growth.sort_values('增长率', ascending=False)

# 5. 构建 product_performance（市场份额 & 利润）
product_performance = df.groupby(['主类别', '子类别'])['美元销售额'].sum().reset_index()
total_sales = product_performance['美元销售额'].sum()
product_performance['sales_share'] = (product_performance['美元销售额'] / total_sales) * 100
product_performance = product_performance.rename(columns={'美元销售额': 'sales'})

# 合并利润
profit_summary = df.groupby(['主类别', '子类别'])['利润'].sum().reset_index()
product_performance = product_performance.merge(profit_summary, on=['主类别', '子类别'])

# 6. 合并用于绘图的数据
plot_data = pd.merge(
    product_performance,
    product_growth_trend,
    on=['主类别', '子类别'],
    how='inner'
)

# 7. 绘制产品创新矩阵
plt.figure(figsize=(14, 8))
scatter = sns.scatterplot(
    data=plot_data,
    x='sales_share',
    y='增长率',
    hue='主类别',
    size='利润',  # 使用 '利润' 而非 'profit'（确保列名一致）
    sizes=(50, 500),
    alpha=0.7
)

plt.title('产品创新矩阵 (市场份额 vs 增长率)')
plt.xlabel('市场份额 (%)')
plt.ylabel('2023年增长率')
plt.axvline(x=5, color='gray', linestyle='--', linewidth=1)  # 高/低份额分界
plt.axhline(y=0.3, color='gray', linestyle='--', linewidth=1)  # 高/低增长分界

# 添加象限标签
plt.text(2, 0.1, '潜力产品', fontsize=12, ha='center', color='black', alpha=0.7)
plt.text(2, 0.4, '新兴明星', fontsize=12, ha='center', color='darkgreen')
plt.text(8, 0.1, '成熟产品', fontsize=12, ha='center', color='black', alpha=0.7)
plt.text(8, 0.4, '主导产品', fontsize=12, ha='center', color='darkblue')

# 8. 标注高增长产品
for _, row in high_growth.iterrows():
    # 在 plot_data 中查找对应 sales_share
    match = plot_data[
        (plot_data['子类别'] == row['子类别']) &
        (plot_data['主类别'] == row['主类别'])
    ]
    if not match.empty:
        x = match['sales_share'].values[0]
        y = row['增长率']
        plt.annotate(
            row['子类别'],
            (x, y),
            xytext=(5, 5),
            textcoords='offset points',
            fontsize=9,
            alpha=0.9
        )

plt.grid(alpha=0.3)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('产品创新矩阵.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 4. 供应链优化建议（关税和运费敏感度分析）
def calculate_optimization(country, category):
    """计算供应链优化潜力"""
    subset = df[(df['国家'] == country) & (df['主类别'] == category)]
    current_profit = subset['利润'].sum()
    
    # 场景1：降低关税10%（通过自由贸易协定）
    tariff_reduction = subset['关税成本'] * 0.1
    scenario1_profit = current_profit + tariff_reduction.sum()
    
    # 场景2：降低运费15%（通过本地仓库）
    shipping_reduction = subset['运费'] * 0.15
    scenario2_profit = current_profit + shipping_reduction.sum()
    
    # 场景3：组合优化
    scenario3_profit = current_profit + tariff_reduction.sum() + shipping_reduction.sum()
    
    return {
        '当前利润': current_profit,
        '降关税利润': scenario1_profit,
        '降运费利润': scenario2_profit,
        '组合优化利润': scenario3_profit,
        '优化潜力': (scenario3_profit - current_profit) / current_profit
    }
```

```python
# ====================== 2. 季节性影响分析 ======================
## ====================== 2. 季节性影响分析 ======================
#分析销售季节性模式，并对典型国家-品类组合进行季节性分解
# 全球季节性模式
# 确保日期列为 datetime 类型
df['日期'] = pd.to_datetime(df['日期'])

# 1. 全球季节性模式（按月平均每年的月度总销售额）
# 先按年-月聚合销售额
df['年份'] = df['日期'].dt.year
df['月份'] = df['日期'].dt.month

monthly_sales = df.groupby(['年份', '月份', '主类别'])['美元销售额'].sum().reset_index()

# 计算每个“月份”的平均销售额（跨年平均）
seasonal_global = monthly_sales.groupby(['月份', '主类别'])['美元销售额'].mean().unstack()

# 绘图
plt.figure(figsize=(12, 8))
seasonal_global.plot(marker='o')
plt.title('全球产品类别季节性销售模式（年均月销售额）')
plt.xlabel('月份')
plt.ylabel('平均月销售额 (美元)')
plt.xticks(range(1, 13))
plt.grid(True, alpha=0.3)
plt.legend(title='产品类别', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('全球季节性.png', dpi=300, bbox_inches='tight')
plt.show()


# 2. 代表性国家季节性对比
def plot_country_seasonality(country):
    country_data = df[df['国家'] == country].copy()
    country_data['月份'] = country_data['日期'].dt.month
    country_data['年份'] = country_data['日期'].dt.year

    # 按年-月聚合，再按月平均
    monthly = country_data.groupby(['年份', '月份'])['美元销售额'].sum().reset_index()
    seasonal = monthly.groupby('月份')['美元销售额'].mean()  # 跨年平均每月销售额

    plt.figure(figsize=(10, 6))
    seasonal.plot(marker='o', color='royalblue')
    plt.title(f'{country} 月度销售季节性（年均）')
    plt.xlabel('月份')
    plt.ylabel('平均月销售额 (美元)')
    plt.xticks(range(1, 13))
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# 分析南北半球代表性国家
plot_country_seasonality('美国')      # 北半球
plot_country_seasonality('澳大利亚')  # 南半球


# 3. 季节性分解示例
from statsmodels.tsa.seasonal import seasonal_decompose

def plot_seasonal_decomposition(country, category):
    data = df[(df['国家'] == country) & (df['主类别'] == category)].copy()
    
    # 设置日期为索引并按月汇总
    data.set_index('日期', inplace=True)
    monthly = data['美元销售额'].resample('M').sum()  # 按月求和
    
    # 检查数据长度（至少24个月）
    if len(monthly) < 24:
        print(f"数据不足，无法分解: {country}-{category}，仅有 {len(monthly)} 个月数据")
        return
    
    try:
        decomposition = seasonal_decompose(monthly, model='multiplicative', period=12)
        
        fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 10))
        decomposition.observed.plot(ax=ax1, title=f'{country}-{category} 销售额原始数据')
        decomposition.trend.plot(ax=ax2, title='趋势成分')
        decomposition.seasonal.plot(ax=ax3, title='季节性成分')
        decomposition.resid.plot(ax=ax4, title='残差成分')
        
        plt.tight_layout()
        plt.savefig(f'季节性分解_{country}_{category}.png', dpi=300, bbox_inches='tight')
        plt.show()
    except Exception as e:
        print(f"季节性分解失败: {country}-{category}, 错误: {e}")

# 示例分析
plot_seasonal_decomposition('美国', '电子产品')
plot_seasonal_decomposition('巴西', '服装')
```
